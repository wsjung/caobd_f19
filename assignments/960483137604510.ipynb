```python
import re, os
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph(name='Ex 6.2.2')

       
for faction in ["heroes", "ambiguous", "villains"]:
    faction_chars = [c[:-4] for c in os.listdir("../data/%s" % faction)]
    for char in faction_chars:
        break
        
        with open("../data/%s/%s.txt" % (faction, char)) as fp:
            markup = fp.read()
        
        # get the character partners
        partners = re.findall(r"\| *partners[\w\W]+?\|.+=", markup)
        # loop over partners
        for partner in partners:
            # for each partner create link between char and partner
            G.add_node(*partners)
            G.add_edge(*partners)

print(nx.info(G))
nx.draw(G)
plt.subplot(121)
nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
```
